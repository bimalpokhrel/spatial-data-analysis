```python
# importing panda
import geopandas as gdp
```

```python
# file upload
in_json=r"/content/data/municipality.json"
in_shp=r"/content/data/municipality.shp"
```

```python
# read geodata frame
gdf=gdp.read_file(in_json)
# gdf=gdp.read_file(in_shp)
gdf.head(8)
```

```python
import matplotlib.pyplot as plt
fig, ax=plt.subplots(1,figsize=(8,10))
gdf.plot(ax=ax,column="Province",legend=True,legend_kwds={"loc":'upper right'})
# lgd=ax.get_legend()
# lgd.set_bbox_to_anchor((1.04,0.5))
```

```python
# adjusting legend to centre
import matplotlib.pyplot as plt
fig, ax=plt.subplots(1,figsize=(5,10))
gdf.plot(ax=ax,column="Province",legend=True,legend_kwds={"loc":'center left'})
lgd=ax.get_legend()
lgd.set_bbox_to_anchor((1.04,0.5))
```

```python
# write vector data
gdf.to_file(r"/content/output/out_shp.shp",encoding="UTF-8")
gdf.to_file(r"//content/output/out.geojson",driver="GeoJSON",encoding="UTF-8")
```
